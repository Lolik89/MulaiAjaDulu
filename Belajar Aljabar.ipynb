```python
import numpy as np
import matplotlib.pyplot as plt

# Persamaan pertama: y = 15x
# Persamaan kedua: -12x + 2y = 4

# Membuat array x dari -10 hingga 10
x = np.linspace(-10, 10, 100)

# Menggunakan persamaan pertama untuk mendapatkan nilai y
y1 = 15 * x

# Menggunakan persamaan kedua untuk mendapatkan nilai y
# (-12x + 2y = 4) diubah menjadi (y = (12x + 4) / 2)
y2 = (12 * x + 4) / 2

# Plot persamaan pertama dan kedua
plt.plot(x, y1, label='y = 15x')
plt.plot(x, y2, label='-12x + 2y = 4')

# Menandai titik potong (solusi)
solusi_x = 2/9
solusi_y = 10/3
plt.scatter(solusi_x, solusi_y, color='red', label=f'Solusi (x={solusi_x}, y={solusi_y})')

# Menambahkan label dan legenda
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Menampilkan plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Persamaan pertama: 3x - 2y = 20
# Persamaan kedua: 2x + 6y = 10

# Membuat array x dari -10 hingga 10
x = np.linspace(-10, 10, 100)

# Menggunakan persamaan pertama untuk mendapatkan nilai y
y1 = (3*x - 20)/2

# Menggunakan persamaan kedua untuk mendapatkan nilai y
y2 = (10 - 2*x)/6

# Plot persamaan pertama dan kedua
plt.plot(x, y1, label='3x - 2y = 20')
plt.plot(x, y2, label='2x + 6y = 10')

# Menandai titik potong (solusi)
solusi_x = 70/11
solusi_y = -100/11
plt.scatter(solusi_x, solusi_y, color='red', label=f'Solusi (x={solusi_x}, y={solusi_y})')

# Menambahkan label dan legenda
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Menampilkan plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi
def f(x):
    return 2*x**2 + 3*x + 1

# Membuat array x
x_values = np.linspace(-10,10,100)

# Menghitung nilai fungsi pada setiap nilai x
y_values = f(x_values)

# Plot fungsi
plt.plot(x_values, y_values, label='2x^2 + 3x + 1')

# Menambahkan garis vertikal untuk menunjukkan x = 3
plt.axvline(x=3, color='red', linestyle='--', label='x = 3')

# Menambahkan label dan legenda
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# Menampilkan plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi
def f(x):
    return (x**2 - 9) / (x - 3)

# Membuat array x dari 2 hingga 4 dengan langkah 0.1 (menghindari nilai 3)
x_values = np.concatenate((np.arange(2, 3, 0.1), np.arange(3.1, 4.1, 0.1)))

# Menghitung nilai fungsi pada setiap nilai x
y_values = f(x_values)

# Plot fungsi
plt.plot(x_values, y_values, label='(x^2 - 9) / (x - 3)')

# Menambahkan garis vertikal untuk menunjukkan x = 3
plt.axvline(x=3, color='red', linestyle='--', label='x = 3')

# Menambahkan label dan legenda
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Limit as x approaches 3 of (x^2 - 9) / (x - 3)')
plt.legend()

# Menampilkan plot
plt.show()
```
