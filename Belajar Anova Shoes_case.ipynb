```python
import numpy as np
import scipy.stats as st
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import io
import pandas as pd
df = pd.read_csv(io.BytesIO(uploaded['Data Latihan ANOVA dan Pearson Correlation (Shoes).csv']))
df = df.dropna()
```

```python
shoes = df[df.category =='shoes']
```

```python
t = shoes.current_price.to_numpy()/1000
b = shoes.raw_price.to_numpy()/1000
```

```python
st.f_oneway(t,b)
```

```python
x = shoes.current_price
y = shoes.raw_price

sns.set_style('darkgrid')
_ = sns.scatterplot (x=x, y=y)
```

```python
st.pearsonr(x,y)
```
