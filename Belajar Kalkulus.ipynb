```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# SOAL 1
def f(x):
    return 2*x**2 + 3*x + 1

# Membuat array x
x = np.linspace(-10,10,100)

# Menghitung nilai fungsi pada setiap nilai x
y = f(x)

# Plot fungsi
plt.plot(x, y, label='2x^2 + 3x + 1')

plt.axvline(x=3, color='red', linestyle='--', label='x = 3')

# Label dan Legenda
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.show()
```

```python
# LATIHAN KALKULUS INTEGRAL

from scipy.integrate import quad
```

```python
# SOAL INTEGRAL 1
# ∫_{3}^{6} 2x/6 dx

def f(x):
    return 2*x / 6

quad(f, 3, 6)
```

```python
# SOAL INTEGRAL 2
# ∫_{2}^{5} (2x^2-5)/6x dx

def g(x):
    return ((2*x**2) - 5) / (6*x)

quad(g,2,5)
```

```python
# SOAL INTEGRAL 32
# ∫_{1}^{3} (7x^3)/12x dx

def h(x):
    return (7 * x**3) / (12 * x)
quad(h,1,3)
```

```python
from sympy import symbols, integrate

# Mendefinisikan simbol
x = symbols('x')

# Mendefinisikan fungsi
f_x = 9 * x**2

# Menghitung integral dari 9x^2
integral_9x2 = integrate(f_x, x)

# Menghitung integral dari x^2
integral_x2 = integrate(x**2, x)

# Menampilkan hasil
print(f'Integral dari 9x^2 adalah: {integral_9x2}')
print(f'9 dikali integral dari x^2 adalah: {9 * integral_x2}')
```

```python
from sympy import symbols, integrate

# Mendefinisikan simbol
x = symbols('x')

# Mendefinisikan fungsi
f_x = x**2 + 2*x**3

# Menghitung integral dari (x^2 + 2x^3)
integral_result = integrate(f_x, x)

# Menampilkan hasil
print(f'Integral dari (x^2 + 2x^3) adalah: {integral_result}')
```

```python
from sympy import symbols, integrate

# Mendefinisikan simbol dan batas integrasi
x, a, b = symbols('x a b')

# Mendefinisikan fungsi
f_x = x**2 + 2*x**3

# Menghitung integral definitif dari (x^2 + 2x^3) dengan batas a dan b
integral_definite_result = integrate(f_x, (x, a, b))

# Menampilkan hasil
print(f'Integral definitif dari (x^2 + 2x^3) dengan batas {a} dan {b} adalah: {integral_definite_result}')
```
