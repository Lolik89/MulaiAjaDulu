LATIHAN SOAL INTEGRAL
- ∫_{3}^{6} 2x/6 dx
- ∫_{2}^{5} (2x^2-5)/6x dx
- ∫_{1}^{3} (7x^3/12x) dx

```python
# SOAL 1
# ∫_{3}^{6} 2x/6 dx

from scipy import integrate

# Fungsi yang akan diintegrasikan
def f(x):
    return 2*x / 6

# Batas bawah dan batas atas
a = 3
b = 6

# Menghitung nilai integral menggunakan quad dari scipy
result, error = integrate.quad(f, a, b)

# Menghitung nilai AUC (Area Under the Curve)
auc = abs(result)

# Menampilkan hasil
print(f'Nilai Integral: {result}')
print(f'Nilai AUC: {auc}')
print(f'Error: {error}')
```

```python
# SOAL 2
# ∫_{2}^{5} (2x^2-5)/6x dx

from scipy import integrate

# Fungsi yang akan diintegrasikan
def f(x):
    return (2*x**2 - 5) / (6*x)

# Batas bawah dan batas atas
a = 2
b = 5

# Menghitung nilai integral menggunakan quad dari scipy
result, error = integrate.quad(f, a, b)

# Menghitung nilai AUC (Area Under the Curve)
auc = abs(result)

# Menampilkan hasil
print(f'Nilai Integral: {result}')
print(f'Nilai AUC: {auc}')
print(f'Error: {error}')
```

```python
# SOAL 3
# ∫_{1}^{3} (7x^3/12x) dx


from scipy import integrate

# Fungsi yang akan diintegrasikan
def f(x):
    return (7 * x**3) / (12 * x)

# Batas bawah dan batas atas
a = 1
b = 3

# Menghitung nilai integral menggunakan quad dari scipy
result, error = integrate.quad(f, a, b)

# Menghitung nilai AUC (Area Under the Curve)
auc = abs(result)

# Menampilkan hasil
print(f'Nilai Integral: {result}')
print(f'Nilai AUC: {auc}')
print(f'Error: {error}')
```

LATIHAN PERKALIAN KALKULUS INTEGRAL
- ∫ c f(x) dx = c ∫f(x) dx + c
- ∫ 9x^2 dx = 9 ∫ x^2 dx
- ∫ (x^2 + 2x^3)

```python
# SOAL 1

from sympy import symbols, integrate

# Mendefinisikan simbol dan konstanta
x, c = symbols('x c')

# Mendefinisikan fungsi f(x)
f_x = x**2  # Gantilah dengan fungsi yang sesuai

# Menghitung integral c * f(x)
integral_result = integrate(c * f_x, x)

# Menampilkan hasil
print(f'Integral dari c * f(x) adalah: {integral_result}')
```

```python
# Menghitung integral c * f(x) + c
integral_result_with_constant = integrate(c * f_x, x) + c

# Menampilkan hasil
print(f'Integral dari c * f(x) + c adalah: {integral_result_with_constant}')
```

```python
# SOAL 2

# Mendefinisikan simbol
x = symbols('x')

# Mendefinisikan fungsi
f_x = 9 * x**2

# Menghitung integral dari 9x^2
integral_9x2 = integrate(f_x, x)

# Menghitung integral dari x^2
integral_x2 = integrate(x**2, x)

# Menampilkan hasil
print(f'Integral dari 9x^2 adalah: {integral_9x2}')
print(f'9 dikali integral dari x^2 adalah: {9 * integral_x2}')
```

```python
# SOAL 3

# Mendefinisikan simbol
x = symbols('x')

# Mendefinisikan fungsi
f_x = x**2 + 2*x**3

# Menghitung integral dari (x^2 + 2x^3)
integral_result = integrate(f_x, x)

# Menampilkan hasil
print(f'Integral dari (x^2 + 2x^3) adalah: {integral_result}')
```

```python
# Mendefinisikan simbol dan batas integrasi
x, a, b = symbols('x a b')

# Mendefinisikan fungsi
f_x = x**2 + 2*x**3

# Menghitung integral definitif dari (x^2 + 2x^3) dengan batas a dan b
integral_definite_result = integrate(f_x, (x, a, b))

# Menampilkan hasil
print(f'Integral definitif dari (x^2 + 2x^3) dengan batas {a} dan {b} adalah: {integral_definite_result}')
```
