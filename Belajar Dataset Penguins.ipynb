Variasi berat badan penguin (weight), berdasarkan jenis kelamin penguin (sex), yang terbagi menjadi female dan male.

```python
f = adelie[adelie.sex == 'Female']['body_mass_g'].to_numpy()/1000
f
```

```python
m = adelie[adelie.sex == 'Male']['body_mass_g'].to_numpy()/1000
m
```

Mean dari data berat badan penguin berdasarkan jenis kelamin female dan male

```python
fbar =f.mean()
fbar
```

```python
mbar =m.mean()
mbar
```

Menghitung variasi berat badan penguin berdasarkan jenis kelamin.

```python
sf = f.var(ddof=1)
sm = m.var(ddof=1)

nf = f.size
nm = m.size

t = (fbar-mbar)/(sf/nf+sm/nm)**(1/2)
t
```

Hitung derajat kebebasan p-value pakai Welch-Satterthwaite

```python
def ws_eqn(sx,sy,nx,ny):
  return (sx/nx + sy/ny)**2 / (sx**2/(nx-1)) +sy**2/(ny**2 * (ny-1))
```
